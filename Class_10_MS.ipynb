```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Line Plot:

x= [1,2,3,4,5]
y= [2,4,6,8,10]

plt.plot(x,y)
plt.title("Example Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Bar Plot

categories= ['A','B','C','D']
values= [3,7,1,5]

plt.bar(categories,values)
plt.title('Bar Plot')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Histogram

data=np.random.randn(1000)
plt.hist(data,bins=100)
plt.title("Hostogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter Plot

x= [5,7,8,7,2,17,2,9,4,11]
y= [99,86,87,88,100,86,103,87,94,78]

plt.scatter(x,y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Pie Chart

Labels= ['Electronics','Fashion','Home Appliances','Groceries']
Sizes= [30,25,20,25]

plt.pie(Sizes,labels=Labels,autopct='%1.1f%%',startangle=90,explode=[0.1,0,0,0])
plt.title("Pie Chart")
plt.show()
```

```python
# Box Plot

data= [np.random.normal(100,10,200),np.random.normal(90,20,200),np.random.normal(80,30,200)]

plt.boxplot(data,patch_artist=True)
plt.title('Box Plot Example')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Heat Map

data= np.random.rand(10,10)

plt.imshow(data, cmap='coolwarm',interpolation='nearest')
plt.title('Heatmap Example')
plt.colorbar()
plt.show()
```
