```python
import pandas as pd

# Sample data

data= {'Student': ['Alice', 'Bob','Charlie','David','Eve'],
'Exam_Score': [85,90,None,88,None]}

df = pd.DataFrame(data)
print(df)
```

```python
# Identifying missing data
#print(df.isnull())
print(df.isnull().sum())
```

```python
# Imputing with mean
df['Exam_Score_Mean']= df['Exam_Score'].fillna(df['Exam_Score'].mean())

#Imputing with median
df['Exam_Score_Median']= df['Exam_Score'].fillna(df['Exam_Score'].median())

#Imputing with mode
df['Exam_Score_Mode']= df['Exam_Score'].fillna(df['Exam_Score'].mode()[0])

print(df)
```

```python
from sklearn.impute import KNNImputer

# KNN Imputer
imputer = KNNImputer(n_neighbors=2)
df_knn= pd.DataFrame(imputer.fit_transform(df[['Exam_Score']]), columns=['Exam_Score'])

print(df_knn)                
```
